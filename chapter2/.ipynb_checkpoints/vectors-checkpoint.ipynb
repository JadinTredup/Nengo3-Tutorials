# Representing a Vector

In nengo, ensembles of neurons represent information. In this model, you will represent a two-dimensional vector with a single ensemble of LIF (leaky integrate-and-fire) neurons. Though this is a case of two-dimensional vector representation, but the ideas naturally generalize to any dimension.

```python
#Setup the envirnment
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

import nengo
from nengo.dists import Uniform
```

## Create the Model

This model has paramenters as described in the book. Since the default max rates in nengo2 are (200, 400), so you have to explicitly specify them to be (100, 200) to create the same model as in the book. However, the default values of tauRC, tauRef, intercepts, radius and noise in nengo2 are same as those mentioned in the book.

```python
model = nengo.Network(label='2D Representation')
with model:
    #Input Nodes
    sin = nengo.Node(output=np.sin)
    cos = nengo.Node(output=np.cos)
    
    #Ensemble with 100 LIF neurons which represents a 2-dimensional signal
    x = nengo.Ensemble(100, dimensions=2, max_rates = Uniform(100, 200))
    
    # Connecnting input to ensemble
    # The indices in ensemble 'x' define which dimension the input will project to
    nengo.Connection(sin, x[0])
    nengo.Connection(cos, x[1])
```

## Add Probes to Collect Data

```python
with model:
    sin_probe = nengo.Probe(sin)                   
    cos_probe = nengo.Probe(cos)                   
    x_probe = nengo.Probe(x, synapse=0.01)   # Decoded ouput of the ensemble
```

## Run the model

```python
sim = nengo.Simulator(model)  # Create the simulator
sim.run(5)  # Run it for 5 seconds
```

## Plot the Results

```python
# Plot the inputs to the ensemble
plt.plot(sim.trange(), sim.data[sin_probe], 'r', label="Sine Input")
plt.plot(sim.trange(), sim.data[cos_probe], 'k', label="Cosine Input")
plt.legend(loc="best")
plt.xlabel('time [s]');

# Plot the decoded output of the ensemble
plt.figure()
plt.plot(sim.trange(), sim.data[x_probe], label="Decoded output")
plt.plot(sim.trange(), sim.data[sin_probe], 'r', label="Sine Input")
plt.plot(sim.trange(), sim.data[cos_probe], 'k', label="Cosine Input")
plt.legend(loc="best")
plt.xlabel('time [s]');
```

The graph on the top shows the two dimensional input provided to the ensemble and the bottom graph shows the the decoded estimate of this two dimensional input.
