# Representing a scalar

You can construct and manipulate a population of neurons in nengo. This model shows shows how the activity of neural populations can be thought of as representing a mathematical variable (a scalar value). 

```python
#Setup the environment
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import nengo
from nengo.dists import Uniform
```

## Create the Model

This model has paramenters as described in the book and uses a single population (ensemble) of 100 LIF neurons. Note that the default max rates in nengo2 are (200, 400), so you have to explicitly specify them to be (100, 200) to create the same model as described in the book. Moreover the 'Node Factory' feature of ensemble in the book maps to the 'neuron_type' in nengo2 which is set to LIF by default.  However, the default values of tauRC, tauRef, radius and intercepts in nengo2 are the same as those mentioned in the book.

```python
model = nengo.Network(label='Many Neurons')
with model:
    #Input
    sin = nengo.Node(lambda t: np.sin(16 * t))      #Input with range 1
    #sin = nengo.Node(lambda t: 4*np.sin(16 * t))   #Input with range increased to 4
    
    #Ensemble with 100 LIF neurons  
    x = nengo.Ensemble(100, dimensions=1, max_rates = Uniform(100, 200))
    
    #Connecting input to ensemble
    nengo.Connection(sin, x) 
```

## Add Probes to Collect Data

Anything that is probed will collect the data it produces over time, allowing us to analyze and visualize it later.

```python
with model:
    sin_probe = nengo.Probe(sin)                  # The original input
    x_probe = nengo.Probe(x, synapse=0.01)        # Decoded ouput of the ensemble
    x_spikes = nengo.Probe(x.neurons, 'spikes')   # The raw spikes from the neuron
```

## Run the model

In order to run the model, you have to create a simulator. Then, you can run that simulator over and over again without affecting the original model.

```python
sim = nengo.Simulator(model)   # Create the simulator
sim.run(1)  # Run it for 1 seconds
```

## Step 6: Plot the results

```python
from nengo.utils.matplotlib import rasterplot

# Plot the decoded output of the ensemble
plt.plot(sim.trange(), sim.data[x_probe], label="Decoded Value")
plt.plot(sim.trange(), sim.data[sin_probe], 'r', label="Input")
plt.xlabel('time [s]')
plt.legend(loc="best")
    
# Plot the spiking output of the ensemble
plt.figure()
rasterplot(sim.trange(), sim.data[x_spikes])
plt.title("Spiking Raster")
plt.xlabel('time [s]')
```

The graph on the top shows the the decoded value of the neural spiking (a linearly decoded estimate of the input). The bottom graph shows the spike raster which is the spiking output of the neuron population.

### Increasing the range of Input

You have seen that the population of neurons does a reasonably good job of representing the input. However neurons cannot represent arbitrary values well and you can verify this by increasing the range of the input to 4 ( sin = nengo.Node(lambda t: 4*np.sin(16 * t)) ). You will observe the same saturation effects as described in the book.
