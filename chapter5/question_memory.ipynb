# Question Answering with Memory

This demo implements a form of question answering with memory. Two features (color and shape) will be bound by circular convolution and stored in a memory population. A cue will be used to determine either one of the features by deconvolution.

When you run the network, it will start by binding `RED` and `CIRCLE` for 0.25 seconds and then binding `BLUE` and `SQUARE` for 0.25 seconds. Both bound semantic pointers are stored in a memory population. Then the network is asked with the cue at a later time and it still responds with the correct answer. For example, when the cue is `CIRCLE` the network responds with `RED`.

```python
# Setup the environment
import matplotlib.pyplot as plt
%matplotlib inline

import nengo
from nengo import spa
```

## Create the model
This model has parameters as described in the book. A 'Memory' element is created using the spa module. 'Memory' is capable of storing a vector over time and it uses an integrator network to do so as discussed in the book. 

```python
dim = 20                            # number of dimensions for the semantic pointers
subdims = 1                         # size of ensembles making up the vector
num_neurons = 300.0                 # total neurons
neurons_pd = int(num_neurons/dim)   # number of neurons per dimension
mode = False

model = spa.SPA(label="Question Answering with Memory")

with model:
    # Buffers with 300 neurons each  
    model.A = spa.Buffer(neurons_per_dimension=neurons_pd, dimensions=dim, subdimensions=subdims, direct = mode)        # color
    model.B = spa.Buffer(neurons_per_dimension=neurons_pd, dimensions=dim, subdimensions=subdims, direct = mode)        # shape
    model.D = spa.Memory(neurons_per_dimension=1000, synapse=0.4, dimensions=dim, subdimensions=subdims, direct = mode) # conv
    model.C = spa.Buffer(neurons_per_dimension=neurons_pd, dimensions=dim, subdimensions=subdims, direct = mode)        # cue 
    model.E = spa.Buffer(neurons_per_dimension=neurons_pd, dimensions=dim, subdimensions=subdims, direct = mode)        # output
    
    # Connect the buffers
    cortical_actions = spa.Actions(
        'D = A * B',
        'E = D * ~C'
    )
    model.cortical = spa.Cortical(cortical_actions)  
```

## Provide Input to the Model

The color input presented to the model is `RED` and then `BLUE` for 0.25 seconds each before being turned off. In the same way the shape input is `CIRCLE` and then `SQUARE` for 0.25 seconds each. Thus, the network will bind alternatingly `RED * CIRCLE` and `BLUE * SQUARE` for 0.5 seconds.

The cue for deconvolving bound semantic pointers will be turned off for 0.5 seconds and then starts cycling through `CIRCLE`, `RED`, `SQUARE`, and `BLUE` within each second. 

```python
def color_input(t):
    if t < 0.25:
        return 'RED'
    elif t < 0.5:
        return 'BLUE'
    else:
        return '0'

def shape_input(t):
    if t < 0.25:
        return 'CIRCLE'
    elif t < 0.5:
        return 'SQUARE'
    else:
        return '0'

def cue_input(t):
    if t < 0.5:
        return '0'
    sequence = ['0', 'CIRCLE', 'RED', '0', 'SQUARE', 'BLUE']
    idx = int(((t - 0.5) // (1. / len(sequence))) % len(sequence))
    return sequence[idx]

with model:
    model.inp = spa.Input(A=color_input, B=shape_input, C=cue_input)
```

## Add Probes to Collect Data

```python
with model:
    model.config[nengo.Probe].synapse = nengo.Lowpass(0.03)
    A_probe = nengo.Probe(model.A.state.output)
    B_probe = nengo.Probe(model.B.state.output)
    C_probe = nengo.Probe(model.C.state.output)
    D_probe = nengo.Probe(model.D.state.output)
    E_probe = nengo.Probe(model.E.state.output)
```

## Run the model

```python
sim = nengo.Simulator(model)
sim.run(3.0)
```

## Plot the results

```python
plt.figure(figsize=(10, 10))
vocab = model.get_default_vocab(dim)

plt.subplot(5, 1, 1)
plt.plot(sim.trange(), model.similarity(sim.data, A_probe))
plt.legend(model.get_output_vocab('A').keys, fontsize='x-small')
plt.ylabel("color")

plt.subplot(5, 1, 2)
plt.plot(sim.trange(), model.similarity(sim.data, B_probe))
plt.legend(model.get_output_vocab('B').keys, fontsize='x-small')
plt.ylabel("shape")

plt.subplot(5, 1, 3)
plt.plot(sim.trange(), model.similarity(sim.data, C_probe))
plt.legend(model.get_output_vocab('C').keys, fontsize='x-small')
plt.ylabel("cue")

plt.subplot(5, 1, 4)
for pointer in ['RED * CIRCLE', 'BLUE * SQUARE']:
    plt.plot(sim.trange(), vocab.parse(pointer).dot(sim.data[D_probe].T), label=pointer)
plt.legend(fontsize='x-small')
plt.ylabel("convolved")

plt.subplot(5, 1, 5)
plt.plot(sim.trange(), spa.similarity(sim.data[E_probe], vocab))
plt.legend(model.get_output_vocab('E').keys, fontsize='x-small')
plt.ylabel("Output")
plt.xlabel("time [s]");
```

The plots of `shape`, `color`, and `convolved` show that first `RED * CIRCLE` and then `BLUE * SQUARE` will be loaded into the `convolved` buffer so after 0.5 seconds it represents the superposition `RED * CIRCLE + BLUE * SQUARE`.

The last plot shows that the output is most similar to the semantic pointer bound to the current cue. For example, when `RED` and `CIRCLE` are being convolved and the cue is `CIRCLE`, the output is most similar to `RED`.
