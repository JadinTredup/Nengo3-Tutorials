# Question Answering with Memory

This model shows a form of question answering with memory. You will bind two features (color and shape) by circular convolution and store them in a memory population. Then you will provide a cue to the model at a later time to determine either one of the features by deconvolution. This model exhibits better cognitive ability since the answers to the questions are provided at a later time and not at the same time as the questions themselves. 

```python
# Setup the environment
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

import nengo
from nengo.spa import Vocabulary
from nengo.spa import Memory
```

## Create the model
This model has parameters as described in the book. A 'Memory' element is created using the spa module. 'Memory' is capable of storing a vector over time and it uses an integrator network to do so as discussed in the book. 

When you run the model, it will start by binding `RED` and `CIRCLE` for 0.25 seconds and then binding `BLUE` and `SQUARE` for 0.25 seconds. Both bound semantic pointers are stored in the 'Memory' population. Then you will ask a question from the model with a cue at a later time and it will still respond with the correct answer. For example, when you provide the cue `CIRCLE` the network will respond with `RED`. 

```python
dim=20         # Number of dimensions 
N_input=300    # Number of neurons in population
N_conv=70      # Number of neurons per dimension in bind/unbind populations
N_mem=50       # Number of neurons per dimension in memory population

rng = np.random.RandomState(7)
vocab = Vocabulary(dimensions=dim, rng=rng, max_similarity=0.1)
model = nengo.Network(label='Question Answering with Memory', seed=8)

with model:
    # 
    A = nengo.Ensemble(n_neurons=N_input, dimensions=dim, label='A')
    B = nengo.Ensemble(n_neurons=N_input, dimensions=dim, label='B')
    C = nengo.Ensemble(n_neurons=N_input, dimensions=dim, label='C')
    D = nengo.Ensemble(n_neurons=N_input, dimensions=dim, label='D')
    E = nengo.Ensemble(n_neurons=N_input, dimensions=dim, label='E')
    
    # Creating an integrator/memory
    memory = nengo.Ensemble(n_neurons=N_mem*dim, dimensions=dim, label='Memory')
    nengo.Connection(memory, memory, synapse=0.4)
    nengo.Connection(D, memory, synapse=0.01)
    
    BIND = nengo.networks.CircularConvolution(n_neurons=N_conv*dim, dimensions=dim)
    nengo.Connection(A, BIND.A)
    nengo.Connection(B, BIND.B)
    nengo.Connection(BIND.output, D, synapse=0.01) 
    
    UNBIND = nengo.networks.CircularConvolution(n_neurons=N_conv*dim, dimensions=dim, invert_a=True)
    nengo.Connection(C, UNBIND.A)
    nengo.Connection(memory, UNBIND.B)
    nengo.Connection(UNBIND.output, E)
    
    CIRCLE=vocab.parse('CIRCLE').v
    BLUE=vocab.parse('BLUE').v
    RED=vocab.parse('RED').v
    SQUARE=vocab.parse('SQUARE').v
    ZERO=[0]*dim
    
    def color_input(t):
        if t < 0.25:
            return RED
        elif t < 0.5:
            return BLUE
        else:
            return ZERO

    def shape_input(t):
        if t < 0.25:
            return CIRCLE
        elif t < 0.5:
            return SQUARE
        else:
            return ZERO

    def cue_input(t):
        if t < 0.5:
            return ZERO
        sequence = [ZERO, CIRCLE, RED, ZERO, SQUARE, BLUE]
        idx = int(((t - 0.5) // (1. / len(sequence))) % len(sequence))
        return sequence[idx]

    
    inputA = nengo.Node(output=color_input, size_out=dim)
    inputB = nengo.Node(output=shape_input, size_out=dim)
    inputC = nengo.Node(output=cue_input, size_out=dim)
    
    nengo.Connection(inputA, A)
    nengo.Connection(inputB, B)
    nengo.Connection(inputC, C)
```

## Provide Input to the Model

The color input presented to the model is `RED` and then `BLUE` for 0.25 seconds each before being turned off. In the same way the shape input is `CIRCLE` and then `SQUARE` for 0.25 seconds each. Thus, the network will bind alternatingly `RED * CIRCLE` and `BLUE * SQUARE` for 0.5 seconds.

The cue for deconvolving bound semantic pointers will be turned off for 0.5 seconds and then starts cycling through `CIRCLE`, `RED`, `SQUARE`, and `BLUE` within each second. 

## Add Probes to Collect Data

```python
with model:
    A_probe = nengo.Probe(A, synapse=0.1)
    B_probe = nengo.Probe(B, synapse=0.1)
    C_probe = nengo.Probe(C, synapse=0.1)
    D_probe = nengo.Probe(D, synapse=0.1)
    E_probe = nengo.Probe(E, synapse=0.5)
    memory_probe = nengo.Probe(memory, synapse=0.5)
```

## Run the model

```python
sim = nengo.Simulator(model)
sim.run(3.0)
```

## Plot the results

```python
plt.figure(figsize=(10, 10))

plt.subplot(5, 1, 1)
plt.plot(sim.trange(), nengo.spa.similarity(sim.data[A_probe], vocab))
plt.legend(vocab.keys, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.0, fontsize=10)
plt.ylabel("color")

plt.subplot(5, 1, 2)
plt.plot(sim.trange(), nengo.spa.similarity(sim.data[B_probe], vocab))
plt.legend(vocab.keys, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.0, fontsize=10)
plt.ylabel("shape")

plt.subplot(5, 1, 3)
plt.plot(sim.trange(), nengo.spa.similarity(sim.data[C_probe], vocab))
plt.legend(vocab.keys, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.0, fontsize=10)
plt.ylabel("cue")

plt.subplot(5, 1, 4)
for pointer in ['RED * CIRCLE', 'BLUE * SQUARE']:
    plt.plot(sim.trange(), vocab.parse(pointer).dot(sim.data[D_probe].T), label=pointer)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.0,fontsize=10)
plt.ylabel("bound")

plt.subplot(5, 1, 5)
plt.plot(sim.trange(), nengo.spa.similarity(sim.data[E_probe], vocab))
plt.legend(vocab.keys, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.0, fontsize=10)
plt.ylabel("Output")
plt.xlabel("time [s]");

plt.figure(figsize=(10, 10))
plt.subplot(5, 1, 1)
for pointer in ['RED * CIRCLE', 'BLUE * SQUARE']:
    plt.plot(sim.trange(), vocab.parse(pointer).dot(sim.data[memory_probe].T), label=pointer)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.0,fontsize=10)
plt.ylabel("memory")
```

The graphs of `shape`, `color`, and `bound` show that first `RED * CIRCLE` and then `BLUE * SQUARE` will be convolved (bound) and loaded into the Memory '`D`' so after 0.5 seconds it represents the superposition `RED * CIRCLE + BLUE * SQUARE`.

The last plot shows that the output is most similar to the semantic pointer bound to the current cue. For example, when `RED` and `CIRCLE` are being bound and the cue is `CIRCLE`, the output is most similar to `RED`.
