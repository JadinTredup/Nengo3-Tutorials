# Question Answering

This model shows a simple form of question answering. The goal of this model is to build a network that will output answers to questions based on supplied statements as described in the book. You will bind two features (color and shape) by circular convolution. Then you will provide a cue to the model to determine either one of the features by deconvolution.

```python
# Setup the environment
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

import nengo
from nengo.spa import Vocabulary
```

## Create the model
This model has parameters as described in the book. You will use the SPA (Semantic Pointer Architecture) to implement this model. Instead of ensembles (mentioned in the book), buffers are created using the spa module, however their underlying functionality is the similar to that of ensembles (capable of representing vectors with no memory). A buffer is a minimal SPA module which is useful for passing data along and is made up of ensembles at the lowest level. Moreover, with SPA, you do not need to create a bind and an unbind network as discussed in the book, since you can use 'Actions' class in spa to directly form hardwired connections between two SPA components, with built in bind and unbind networks as shown in the code.

When you run the network, it will start by binding `RED` and `CIRCLE` for 0.5 seconds and then binding `BLUE` and `SQUARE` for 0.5 seconds. In parallel you will ask a question from the model by providing a cue and it will provide the answers at the same time as the questions. For example, when the cue is `CIRCLE` the model will respond with `RED`.

Depending on your computer, this model may run slowly which can be avoided by several ways as discussed in the book. In order to run the buffers in direct mode (explained in the book), the parameter 'direct' should be set to True while creating the buffers.

```python

dim=20         # Number of dimensions 
N_input=300    # Number of neurons in population
N_conv=70      # Number of neurons per dimension in bind/unbind populations

rng = np.random.RandomState(7)
vocab = Vocabulary(dimensions=dim, rng=rng, max_similarity=0.1)
model = nengo.Network(label='Question Answering', seed=7)

with model:
    # 
    A = nengo.Ensemble(n_neurons=N_input, dimensions=dim, label='A')
    B = nengo.Ensemble(n_neurons=N_input, dimensions=dim, label='B')
    C = nengo.Ensemble(n_neurons=N_input, dimensions=dim, label='C')
    D = nengo.Ensemble(n_neurons=N_input, dimensions=dim, label='D')
    E = nengo.Ensemble(n_neurons=N_input, dimensions=dim, label='E') 
    
    BIND = nengo.networks.CircularConvolution(n_neurons=N_conv*dim, dimensions=dim)
    nengo.Connection(A, BIND.A)
    nengo.Connection(B, BIND.B)
    nengo.Connection(BIND.output, D) 
    
    UNBIND = nengo.networks.CircularConvolution(n_neurons=N_conv*dim, dimensions=dim, invert_a=True)
    nengo.Connection(C, UNBIND.A)
    nengo.Connection(D, UNBIND.B)
    nengo.Connection(UNBIND.output, E)

    
    CIRCLE=vocab.parse('CIRCLE').v
    BLUE=vocab.parse('BLUE').v
    RED=vocab.parse('RED').v
    SQUARE=vocab.parse('SQUARE').v
    ZERO=[0]*dim
    
    def color_input(t):
        if (t // 0.5) % 2 == 0:
            return RED
        else:
            return BLUE

    def shape_input(t):
        if (t // 0.5) % 2 == 0:
            return CIRCLE
        else:
            return SQUARE

    def cue_input(t):
        sequence = [ZERO, CIRCLE, RED, ZERO, SQUARE, BLUE]
        idx = int((t // (1. / len(sequence))) % len(sequence))
        return sequence[idx]
      
    
    inputA = nengo.Node(output=color_input, size_out=dim)
    inputB = nengo.Node(output=shape_input, size_out=dim)
    inputC = nengo.Node(output=cue_input, size_out=dim)
    
    
    nengo.Connection(inputA, A)
    nengo.Connection(inputB, B)
    nengo.Connection(inputC, C)
```

## Provide Input to the Model

As stated in the book, the color input will alternate every 0.5 seconds between `RED` and `BLUE`. In the same way the shape input alternates between `CIRCLE` and `SQUARE`. Thus, the model will bind alternatingly `RED * CIRCLE` and `BLUE * SQUARE` for 0.5 seconds each.

The cue for deconvolving bound semantic pointers cycles through `CIRCLE`, `RED`, `SQUARE`, and `BLUE` within one second. 

## Add Probes to Collect Data

```python
with model:
    A_probe = nengo.Probe(A, synapse=0.1)
    B_probe = nengo.Probe(B, synapse=0.1)
    C_probe = nengo.Probe(C, synapse=0.1)
    D_probe = nengo.Probe(D, synapse=0.1)
    E_probe = nengo.Probe(E, synapse=0.1)
```

## Run the model

```python
sim = nengo.Simulator(model)   # Create the simulator
sim.run(3.0)    # Run it for 3 seconds
```

## Plot the results

```python
plt.figure(figsize=(10, 10))

plt.subplot(5, 1, 1)
plt.plot(sim.trange(), nengo.spa.similarity(sim.data[A_probe], vocab))
plt.legend(vocab.keys, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.0, fontsize=10)
plt.ylabel("color")

plt.subplot(5, 1, 2)
plt.plot(sim.trange(), nengo.spa.similarity(sim.data[B_probe], vocab))
plt.legend(vocab.keys, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.0, fontsize=10)
plt.ylabel("shape")

plt.subplot(5, 1, 3)
plt.plot(sim.trange(), nengo.spa.similarity(sim.data[C_probe], vocab))
plt.legend(vocab.keys, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.0, fontsize=10)
plt.ylabel("cue")

plt.subplot(5, 1, 4)
for pointer in ['RED * CIRCLE', 'BLUE * SQUARE']:
    print pointer
    plt.plot(sim.trange(), vocab.parse(pointer).dot(sim.data[D_probe].T), label=pointer)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.0,fontsize=10)
plt.ylabel("bound")

plt.subplot(5, 1, 5)
plt.plot(sim.trange(), nengo.spa.similarity(sim.data[E_probe], vocab))
plt.legend(vocab.keys, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.0, fontsize=10)
plt.ylabel("Output")
plt.xlabel("time [s]");
```

The last graph shows that the output is most similar to the semantic pointer bound to the current cue. For example, when `RED` and `CIRCLE` are being bound and the cue is `CIRCLE`, the output is most similar to `RED`.
