# Routed Sequencing with Cleanup all Memory

The previous model on Routed Sequencing with cleanup memory used a scalar ensemble to project only state 'A'. In this model, you will project all the states onto an ensemble of neurons as the state cycles through a five element sequence.

```python
# Setup the environment
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np 

import nengo
from nengo import spa                #import spa related packages
from nengo.spa import Vocabulary
```

## Create the model

This model has parameters as described in the book. It thus extends the routed sequencing model by creating a scalar ensemble 'cleanup' and then projecting all the states on to this ensemble using a transformation matrix 'pd_new', specified by the vectors in the vocabulary. 

```python
# Number of dimensions for the Semantic Pointers
dimensions = 16

# Make a model object with the SPA network
model = spa.SPA(label='Routed_Sequence with cleanupAll')

with model:
    # Specify the modules to be used
    model.state = spa.Buffer(dimensions=dimensions)
    model.vision = spa.Buffer(dimensions=dimensions) 
    # Specify the action mapping
    actions = spa.Actions(
        'dot(vision, START) --> state = vision',
        'dot(state, A) --> state = B',
        'dot(state, B) --> state = C',
        'dot(state, C) --> state = D',
        'dot(state, D) --> state = E',
        'dot(state, E) --> state = A'
    )
    
    # Creating the BG and Thalamus components that confirm to the specified rules
    model.BG = spa.BasalGanglia(actions=actions)
    model.thal = spa.Thalamus(model.BG)
    
    # Change the seed of this RNG to change the vocabulary
    rng = np.random.RandomState(0)
    vocab = Vocabulary(dimensions=dimensions, rng=rng)

    # Create the transformation matrix (pd_new) and the cleanup ensemble (cleanup)      
    vsize = len((model.get_output_vocab('state').keys)) - 1

    pd = []
    for item in range(vsize):
        pd.append(model.get_output_vocab('state').keys[item+1])  
    
    pd_new = []
    for element in range(vsize):
        pd_new.append([vocab[pd[element]].v.tolist()]) 
      
    cleanup = nengo.Ensemble(300, dimensions=vsize)
    
    # Function that provides the model with an initial input semantic pointer.
    def start(t):
        if t < 0.05:
            return '0.8*START+D'
        else:
            return '0'

    # Input
    model.input = spa.Input(vision=start)
    
    # Projecting the state on to the cleanup ensemble using a transformation matrix 'pd'.
    for i in range(5):
        nengo.Connection(model.state.state.output, cleanup[i], transform=pd_new[i])
```

## Add Probes to Collect Data

```python
with model:
    state_p = nengo.Probe(model.state.state.output, synapse=0.01)
    vision_p = nengo.Probe(model.vision.state.output, synapse=0.01)
    actions_p = nengo.Probe(model.thal.actions.output, synapse=0.01)
    utility_p = nengo.Probe(model.BG.input, synapse=0.01)
    cleanup_p = nengo.Probe(cleanup, synapse=0.01)
```

## Run the model

```python
sim = nengo.Simulator(model)     # Create the simulator
sim.run(0.5)    # Run it for 0.5 seconds
```

## Plot the results

```python
fig = plt.figure(figsize=(12,8))
p1 = fig.add_subplot(4,1,1)
p1.plot(sim.trange(), model.similarity(sim.data, vision_p))
p1.legend(model.get_output_vocab('vision').keys, fontsize='x-small')
p1.set_ylabel('Vision')

p2 = fig.add_subplot(4,1,2)
p2.plot(sim.trange(), model.similarity(sim.data, state_p))
p2.legend(model.get_output_vocab('state').keys, fontsize='x-small')
p2.set_ylabel('State')

p3 = fig.add_subplot(4,1,3)
p3.plot(sim.trange(), sim.data[cleanup_p], 'k')
p3.set_ylabel('CleanupAll')
p3.set_ylim(-0.6,0.8)

fig = plt.figure(figsize=(12,8))
p4 = fig.add_subplot(4,1,1)
p4.plot(sim.trange(), sim.data[actions_p])
p4_legend_txt = [a.effect for a in model.BG.actions.actions]
p4.legend(p4_legend_txt, fontsize='x-small')
p4.set_ylabel('Action')

p5 = fig.add_subplot(4,1,2)
p5.plot(sim.trange(), sim.data[utility_p])
p5_legend_txt = [a.condition for a in model.BG.actions.actions]
p5.legend(p5_legend_txt, fontsize='x-small')
p5.set_ylabel('Utility')

fig.subplots_adjust(hspace=0.2)
```

Since the cleanup operation is similar to a dot product between the state and the defined vocabulary vectors, the value of the cleanup population in a particular dimension should rise only when the value of 'state' corresponds to that dimension. The response of the cleanup population is shown in the third plot above.
