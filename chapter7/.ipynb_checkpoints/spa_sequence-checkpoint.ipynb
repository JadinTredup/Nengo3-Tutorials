# Sequencing

This model uses the basal ganglia model to cycle through a sequence of five representations (i.e., A->B->C->D->E->A->...). The model incorporates a working memory component (state), which allows the basal ganglia to update that memory based on a set of condition/action mappings.

```python
#Setup the environment
import matplotlib.pyplot as plt
%matplotlib inline

import nengo                
from nengo import spa       #import spa related packages
```

## Create the model

The model has parameters as described in the book. In nengo1.4 (book) separate 'Rules' and 'Sequence' classes were created, however, this is not needed in nengo2 since you can directly specify the rules using the built-in 'Actions' class in spa (semantic pointer architecutre) package. This class takes a string definition of the action as an input as shown in the code where '-->' is used to split the action into condition and effect, otherwise it is treated as having no condition and just effect.

The syntax for creating an input function in nengo2 is also different from that in nengo1.4 mentioned in the book. The syntax for nengo2 which you will use here is spa.input(module=function). The first parameter 'module' refers to name of the module that you want to provide input to and the second parameter 'function' refers to the function to execute to generate inupts to that module. The functions should always return strings, which will then be parsed by the relevant vocabulary.

In nengo1.4, a memory element for representing the `state` was created by using Buffer() as described in the book. However, in nengo2, Buffer() is just module capable of representing an single vector with no memory (similar to an ensemble). Thus, you will have to use Memory() to create a module which is capable of storing a vector over time.  

```python
#Number of dimensions for the Semantic Pointers
dimensions = 16

#Make a model with the SPA network
model = spa.SPA(label='Sequence')
with model:
    #Creating a working memory/cortical element 
    model.state = spa.Memory(dimensions=dimensions)
    
    #Specifying the action mappings (rules) for BG and Thal 
    actions = spa.Actions(
        'dot(state, A) --> state = B',
        'dot(state, B) --> state = C',
        'dot(state, C) --> state = D',
        'dot(state, D) --> state = E',
        'dot(state, E) --> state = A'
    )
    
    #Creating the BG and Thalamus components that confirm to the specified rules
    model.BG = spa.BasalGanglia(actions=actions)    
    model.thal = spa.Thalamus(model.BG)             
    
    #Function that provides the model with an initial input semantic pointer.
    def start(t):
        if t < 0.1:          #Duration of the initial input = 0.1
            return 'D'
        else:
            return '0'
        
    #Input
    model.input = spa.Input(state=start)
```

## Add Probes to Collect Data

```python
with model:
    state_p = nengo.Probe(model.state.state.output, synapse=0.01)
    actions_p = nengo.Probe(model.thal.actions.output, synapse=0.01)
    utility_p = nengo.Probe(model.BG.input, synapse=0.01)
```

## Run the model

```python
sim = nengo.Simulator(model)    # Create the simulator
sim.run(0.5)                    # Run it for 0.5 seconds
```

## Plot the results

```python
fig = plt.figure(figsize=(12,8))
p1 = fig.add_subplot(3,1,1)

p1.plot(sim.trange(), model.similarity(sim.data, state_p))
p1.legend(model.get_output_vocab('state').keys, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.0, fontsize=10)
p1.set_ylabel('State');

p2 = fig.add_subplot(3,1,2)
p2.plot(sim.trange(), sim.data[actions_p])
p2_legend_txt = [a.effect for a in model.BG.actions.actions]
p2.legend(p2_legend_txt, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.0, fontsize=10)
p2.set_ylabel('Action');

p3 = fig.add_subplot(3,1,3)
p3.plot(sim.trange(), sim.data[utility_p])
p3_legend_txt = [a.condition for a in model.BG.actions.actions]
p3.legend(p3_legend_txt, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.0, fontsize=10)
p3.set_ylabel('Utility')
p3.set_xlabel("time [s]");

fig.subplots_adjust(hspace=0.2)
```

The first graph on the top shows the semantic pointer representation of the values stored in the 'state' ensemble. The second graph shows the current transition or the action being executed, and the third graph shows the utility (similarity) of the current basal ganglia input (i.e., state) with the possible vocabulary vectors. 

The book describes that the results of the model can be seen through the visualizer in nengo1.4 GUI which has a "Utility" box and the "Rules" box. Note that the second graph shows the same information as seen in the "Rules" box and third graph shows the same information as seen in the "Utility" box.
