# Learning a communication channel

This model shows how to include synaptic plasticity in nengo models using the hPES learning rule. You will implement error-driven learning to learn to compute a simple communication channel. This is done by using an error signal provided by a neural ensemble to modulate the connection between two other ensembles.

```python
# Setup the environment
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import nengo
from nengo.processes import WhiteNoise
```

## Create the Model

The model has parameters as described in the book. Note that hPES learning rule is built into nengo1.4 as described in the book. However, it can be implemented in nengo2 by combining the PES and the BCM rule as shown. Also, instead of using the 'gate' and 'switch' as described in the book, 'inhibit' population is used which serves the same purpose of turning off the learning by inhibiting the error population. For computing the actual error (which is required only for analysis), direct mode can be used by specifying the neuron_type as 'nengo.Direct()' while creating the 'actual_error' ensemble.


When you run the model, you will see that the 'post' population gradually learns to compute the communication channel. In the model, you will inhibit the error population after 10 seconds to turn off learning and you will see that the 'post' population will still track the 'pre' population showing that the model has actually learned the input. 

```python
model = nengo.Network(label='Learning')
with model:
    # Ensembles to represent populations
    pre = nengo.Ensemble(50, dimensions=1)
    post = nengo.Ensemble(50, dimensions=1)
    error = nengo.Ensemble(100, dimensions=1)
    inhibit = nengo.Ensemble(100, dimensions=1)
    actual_error = nengo.Ensemble(100, dimensions=1, neuron_type=nengo.Direct())
    
    # Actual Error = pre - post (direct mode)
    nengo.Connection(pre, actual_error)
    nengo.Connection(post, actual_error, transform=-1)
    
    # Error = pre - post
    nengo.Connection(pre, error, synapse=0.02)
    nengo.Connection(post, error, transform=-1, synapse=0.02)
    
    # Modulatory connections don't impart current
    error_conn = nengo.Connection(error, post, modulatory=True, synapse=0.02)
    
    # Connecting pre population to post population (communication channel)
    conn = nengo.Connection(pre, post, function=lambda x: np.random.random(1), solver=nengo.solvers.LstsqL2(weights=True))
    
    # Adding the learning rule to the connection
    conn.learning_rule_type = [nengo.PES(error_conn, learning_rate=5e-7), nengo.BCM(learning_rate=1e-11)]
    
    # Providing input to the model
    input = nengo.Node(WhiteNoise(60, high=10).f())     # f() returns a function that can be passed to a node.
    nengo.Connection(input, pre, synapse=0.02)          # Connecting input to the pre ensemble
    
    # Inhibit the error population after 10 seconds
    def inhib(t):
        return 2.0 if t > 10.0 else 0.0
    
    inhibit = nengo.Node(inhib)
    nengo.Connection(inhibit, error.neurons, transform=[[-1]] * error.n_neurons, synapse=0.01)
```

## Add Probes to Collect Data

```python
with model:    
    input_p = nengo.Probe(input, synapse=0.01)
    pre_p = nengo.Probe(pre, synapse=0.01)
    post_p = nengo.Probe(post, synapse=0.01)
    error_p = nengo.Probe(error, synapse=0.01)
    actual_error_p = nengo.Probe(error, synapse=0.01)
```

## Run the Model

```python
sim = nengo.Simulator(model)   # Create the simulator
sim.run(20.0)    # Run it for 20 seconds
```

## Plot The Results

```python
plt.figure(figsize=(12, 12))

plt.subplot(3, 1, 1)
plt.plot(sim.trange(), sim.data[input_p].T[0], c='k', label='Input')
plt.plot(sim.trange(), sim.data[pre_p].T[0], c='b', label='Pre')
plt.plot(sim.trange(), sim.data[post_p].T[0], c='r', label='Post')
plt.ylabel("Dimension 1")
plt.legend(loc='best')

plt.subplot(3, 1, 2)
plt.plot(sim.trange(), sim.data[error_p], c='b', label='Error')
plt.legend(loc='best')
plt.ylim(-1, 1)

plt.subplot(3, 1, 3)
plt.plot(sim.trange(), sim.data[actual_error_p], c='b', label='Actual Error')
plt.legend(loc='best')
plt.ylim(-1, 1)
```

The topmost graph shows how the post population initially doesn't track the pre population but eventually it learns the function (which is a communication channel in this case) and starts tracking the post population. The other two graphs show that there is significant error between the pre and the post populations at the begining which eventually gets reduced to zero as model learns the communication channel.
